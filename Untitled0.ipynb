```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

iname = "/content/drive/MyDrive/Report_ImageProgress_Nhóm8/VuHoaBinh_21106541/download.jpg"

image = cv2.imread(iname, cv2.IMREAD_GRAYSCALE)

if image is None:
 print("Could not open the image")
image
```

• Bô lọc Robert Cross Gradient


```python
kernel_prewitt_x = np.array([[0, 0, 0],
                            [0, -1, 0],
                             [0, 0, 1]])
kernel_prewitt_y = np.array([[0, 0, 0],
                            [0, 0, -1],
                            [0, 1, 0]])

# gradient_x = cv2.filter2D(image, cv2.CV_64F, kernel_prewitt_x, cv2.BORDER_REPLICATE)
gradient_x_robert = cv2.filter2D(image, cv2.CV_64F, kernel_prewitt_x, borderType=cv2.BORDER_REPLICATE)
gradient_y_robert = cv2.filter2D(image, cv2.CV_64F, kernel_prewitt_y, borderType=cv2.BORDER_REPLICATE)
gradient_magnitude = np.sqrt(gradient_x_robert**2 + gradient_y_robert**2)

gradient_magnitude[gradient_magnitude <= 70] = 0

gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)


plt.title("original image")
plt.imshow(image, cmap='gray')
plt.show()

plt.title("Robert X")
plt.imshow(gradient_x_robert, cmap='gray')
plt.show()

plt.title("Robert Y")
plt.imshow(gradient_y_robert, cmap='gray')
plt.show()


plt.title("Processed Image - Robert")
plt.imshow(gradient_magnitude, cmap='gray')
plt.show()

```

```python
## sắc nét
# gradient_magnitude = np.abs(gradient_x_robert) + np.abs(gradient_y_robert)
plt.title("sắc nét")
plt.imshow( image + gradient_magnitude, cmap='gray')

plt.show()
```

• Bô lọc Sobel


```python
kernel_prewitt_x = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
kernel_prewitt_y = np.array([[-1, 0, 1], [-2, 0, 2],[-1, 0, 1]])

# gradient_x = cv2.filter2D(image, cv2.CV_64F, kernel_prewitt_x, cv2.BORDER_REPLICATE)
gradient_x = cv2.filter2D(image, cv2.CV_64F, kernel_prewitt_x, borderType=cv2.BORDER_REPLICATE)
gradient_y = cv2.filter2D(image, cv2.CV_64F, kernel_prewitt_y, borderType=cv2.BORDER_REPLICATE)

gradient_magnitude_x = np.sqrt(gradient_x**2)
gradient_magnitude_y = np.sqrt(gradient_y**2)
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

gradient_magnitude[gradient_magnitude <= 200] = 0

gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

plt.title("original image")
plt.imshow(image, cmap='gray')
plt.show()

plt.title("Robert X")
plt.imshow(gradient_magnitude_x, cmap='gray')
plt.show()

plt.title("Robert Y")
plt.imshow(gradient_magnitude_y, cmap='gray')
plt.show()


plt.title("Processed Image - Robert")
plt.imshow(gradient_magnitude, cmap='gray')
plt.show()
```

```python
## sắc nét
plt.title("sắc nét")
plt.imshow(np.abs(image + gradient_magnitude), cmap='gray')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

iname1 = "/content/drive/MyDrive/Report_ImageProgress_Nhóm8/VuHoaBinh_21106541/1233.jpg"

image1 = cv2.imread(iname1, cv2.IMREAD_GRAYSCALE)

if image1 is None:
 print("Could not open the image")
image1
```

```python
kernel_prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
kernel_prewitt_y = np.array([[1, 1, 1], [0, 0, 0],[-1, -1, -1]])

# gradient_x = cv2.filter2D(image, cv2.CV_64F, kernel_prewitt_x, cv2.BORDER_REPLICATE)
gradient_x = cv2.filter2D(image1, cv2.CV_64F, kernel_prewitt_x, borderType=cv2.BORDER_REPLICATE)
gradient_y = cv2.filter2D(image1, cv2.CV_64F, kernel_prewitt_y, borderType=cv2.BORDER_REPLICATE)

gradient_magnitude_x = np.sqrt(gradient_x**2)
gradient_magnitude_y = np.sqrt(gradient_y**2)
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

gradient_magnitude[gradient_magnitude <= 1] = 0

gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(image1, cmap='gray')
ax[0].set_title("Original Image")

ax[1].imshow(gradient_magnitude, cmap='gray')
ax[1].set_title("Processed Image (Gradient Magnitude)")

plt.show()
```

```python
## sắc nét
gradient_magnitude = np.abs(gradient_x) + np.abs(gradient_y)
plt.title("sắc nét")
plt.imshow(np.abs(image1 + gradient_magnitude), cmap='gray')

plt.show()
```
