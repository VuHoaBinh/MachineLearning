##1. Erosion

```python
import cv2
import numpy as np

img = cv2.imread("/content/drive/MyDrive/Report_ImageProgress_Nhóm8/VuHoaBinh_21106541/Image2/i.jpg",0)
kernel = np.ones((5,5),np.uint8)


erosion = cv2.erode(img,kernel,iterations = 1)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(erosion,cmap = 'gray')
plt.title('Erosion'), plt.xticks([]), plt.yticks([])

plt.show()
```

##2. Dilation

```python
dilation = cv2.dilate(image,kernel,iterations = 1)

plt.subplot(121),plt.imshow(image,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dilation,cmap = 'gray')
plt.title('Dilation'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img = cv2.imread("/content/drive/MyDrive/Report_ImageProgress_Nhóm8/VuHoaBinh_21106541/Image2/i2.jpg",0)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(opening,cmap = 'gray')
plt.title('Opening'), plt.xticks([]), plt.yticks([])

plt.show()
```

## Bài tập trong slides

```python
img = cv2.imread("/content/drive/MyDrive/Report_ImageProgress_Nhóm8/VuHoaBinh_21106541/Image2/Finger.png", cv2. IMREAD_GRAYSCALE)
img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]
img = cv2.bitwise_not(img)
plt.imshow(img, cmap='gray')
plt.show()
```

##Some Morphological Algorithms


```python
# kernel get 8  = B
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

## A erosion B
erosion = cv2.erode(img,kernel,iterations = 1)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(erosion,cmap = 'gray')
plt.title('Erosion'), plt.xticks([]), plt.yticks([])
plt.show()

## ( A erosion B ) dilation B = A opening B
dilation = cv2.dilate(erosion,kernel,iterations = 1)
plt.subplot(122),plt.imshow(dilation,cmap = 'gray')
plt.title('Dilation'), plt.xticks([]), plt.yticks([])
plt.show()

## A opening B
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
plt.subplot(122),plt.imshow(opening,cmap = 'gray')
plt.title('Opening'), plt.xticks([]), plt.yticks([])
plt.show()

```

```python
## (A opening B) dilation B
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
plt.subplot(122),plt.imshow(opening,cmap = 'gray')
plt.title('Opening'), plt.xticks([]), plt.yticks([])
plt.show()

dilation = cv2.dilate(opening,kernel,iterations = 1)
plt.subplot(122),plt.imshow(dilation,cmap = 'gray')
plt.title('Dilation'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
## ((A opening B) dilation B) enrosion B
dilation = cv2.dilate(opening,kernel,iterations = 1)
plt.subplot(122),plt.imshow(dilation,cmap = 'gray')
plt.title('Dilation'), plt.xticks([]), plt.yticks([])
plt.show()


erosion = cv2.erode(dilation,kernel,iterations = 1)
plt.subplot(122),plt.imshow(erosion,cmap = 'gray')
plt.title('Erosion'), plt.xticks([]), plt.yticks([])
plt.show()
```

##Some Morphological Algorithms


```python
img = cv2.imread("/content/drive/MyDrive/Report_ImageProgress_Nhóm8/VuHoaBinh_21106541/Image2/tool.jpg", cv2. IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.show()
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
```

```python
## A dilation B
dilation = cv2.dilate(img,kernel,iterations = 1)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dilation,cmap = 'gray')
plt.title('Dilation'), plt.xticks([]), plt.yticks([])
plt.show()

##(A dilation B) - A
sub1 = cv2.subtract(dilation, img)
plt.subplot(122),plt.imshow(sub1,cmap = 'gray')
plt.title('Subtraction'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
## A erosion B
erosion = cv2.erode(img,kernel,iterations = 1)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(erosion,cmap = 'gray')
plt.title('Erosion'), plt.xticks([]), plt.yticks([])
plt.show()

## A - (A erosion B)
sub2 = cv2.subtract(img, erosion)
plt.subplot(122),plt.imshow(sub2,cmap = 'gray')
plt.title('Subtraction'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
## sub1 + sub2
sub = cv2.add(sub1, sub2)
plt.subplot(122),plt.imshow(sub,cmap = 'gray')
plt.title('Addition'), plt.xticks([]), plt.yticks([])
plt.show()
```
